```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from collections import OrderedDict
import matplotlib




plt.figure(figsize=(20,10));


# def dis(x1,y1,x2,y2):
#     return np.sqrt((np.power(x1-x2,2))+(np.power(y1-y2,2)));


def min_distance_classifier(X,mean):
    return np.dot(X, mean)-0.5*np.dot(mean, mean);

df=pd.read_csv('train.txt',sep=' ',header=None,dtype='Int64')
print(df)
dVal=df.values;




x_count1=0;
y_count1=0;

x_count2=0;
y_count2=0;
i=0;
j=0;
for row in dVal:
    if row[2]==1:
        plt.scatter(row[0],row[1],color='red',marker='.',label='Train Data Of Class 1',s=12**2);
        x_count1=x_count1+row[0];
        y_count1=y_count1+row[1];
        i=i+1;
    if row[2]==2:
                plt.scatter(row[0],row[1],color='green',marker='x',label='Train Data Of Class 2',s=12**2);
                x_count2=x_count2+row[0];
                y_count2=y_count2+row[1];
                j=j+1;

                

mean_x1=x_count1/i;
mean_y1=y_count1/i;


mean_x2=x_count2/j;
mean_y2=y_count2/j;

mean1=np.array([mean_x1,mean_y1]);
mean2=np.array([mean_x2,mean_y2]);

plt.scatter(mean_x1,mean_y1,color='red',marker='v',label='Mean of Class 1',s=12**2);
plt.scatter(mean_x2,mean_y2,color='green',marker='s',label='Mean of Class 2',s=12**2);


test=pd.read_csv('test.txt',sep=' ',header=None,dtype='Int64');
tval=test.values;


cc=0;
for x,y,l in tval:
    arr=np.array([x,y]);
    if min_distance_classifier(arr,mean1)>=min_distance_classifier(arr,mean2):
        plt.scatter(x,y,color='red',marker='o',label='Test data classified as Class 1',s=12**2);
        if l==1:
            cc=cc+1;
                
    else:
        
            
        plt.scatter(x,y,color='green',marker='X',label='Test data classified as Class 2',s=12**2);
        if l==2:
            cc=cc+1;
        
        


m1=mean_x1-mean_x2;
m2=mean_y1-mean_y2;
cons=-0.5*((mean_x1*mean_x1+mean_y1*mean_y1)-(mean_x2*mean_x2+mean_y2*mean_y2));

x=np.linspace(-5,10,100);
y=(m1*x+cons)/(-m2);


plt.plot(x,y,linestyle='-',linewidth=1);



handles, labels = plt.gca().get_legend_handles_labels();
by_label = OrderedDict(zip(labels, handles));
plt.legend(by_label.values(), by_label.keys(),scatterpoints=1,fontsize=15);                
                
    
print("Correctly Classified: ",cc);
print("\nAccuracy=",(cc/len(tval))*100,'%');


                
    
```
